```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import lichen.lichen as lch

import seaborn as sn
```

```python
#a=np.loadtxt('summary_log_likelihoods_with_and_without_WIMP_FIT0001C.dat',dtype='str')
a=np.loadtxt('summary_log_likelihoods_with_and_without_WIMP_FIT0002B.dat',dtype='str')

#tag = 'FIT0001C' # Data has no WIMP
tag = 'FIT0002B' # Data has a WIMP added with M=7 GeV, sigma_n=7e-42. 

#ranges = (-5,5)
ranges = (-20,5)

a = a.transpose()

lh_noWIMP,lh_withWIMP,diff,D,prob = a[1:].astype(float)
```

```python
plt.figure(figsize=(16,4),dpi=100)
plt.subplot(1,3,1)

lch.hist_err(diff,color='black',bins=50,linewidth=2,range=ranges)
plt.xlabel(r'$-\log \mathcal{L}_{\rm no WIMP} - (-\log \mathcal{L}_{\rm with WIMP})$',fontsize=24)
plt.ylabel('# entries',fontsize=18)
#plt.yscale('log')
plt.tight_layout()

#plt.figure()
plt.subplot(1,3,2)
ranges=(2*ranges[0],2*ranges[1])
lch.hist_err(2*diff,color='black',bins=50,linewidth=2,range=ranges)
plt.xlabel(r'$D = 2 \times \left[\Delta(-\log \mathcal{L})\right]$',fontsize=24)
plt.ylabel('# entries',fontsize=18)
plt.tight_layout()

plt.subplot(1,3,3)
lch.hist_err(prob,color='black',bins=50,linewidth=2)
plt.xlabel(r'$\chi^2$ probability density function',fontsize=24)
plt.ylabel('# entries',fontsize=18)
#plt.ylim(0,60)
plt.tight_layout()

name = "chisq_summary_%s.png" % (tag)
plt.savefig(name)

```

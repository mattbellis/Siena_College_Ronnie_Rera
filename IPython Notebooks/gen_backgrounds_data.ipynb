```python
# Test the surface events

e = np.linspace(0.5,4.0,1000)
t = np.linspace(0,1000,1000)
rt = np.linspace(0,7.0)

# Using Nicole's simulated stuff
mu = [0.269108,0.747275,0.068146]
sigma = [0.531530,-0.020523]


ye = 0.0806*e*e + -0.5125*e + 1
yrt = rise_time_prob_exp_progression(rt,4.0*np.ones(len(rt)),mu,sigma,0,7)
yt = np.exp(-0.0002*t)

plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.plot(e,ye)
plt.subplot(1,3,2)
plt.plot(t,yt)
plt.subplot(1,3,3)
plt.plot(rt,yrt)

```

```python
def gen_surface_events(maxpts,max_days,name_of_output_file):
    
    # Using Nicole's simulated stuff
    mu = [0.269108,0.747275,0.068146]
    sigma = [0.531530,-0.020523]


    max_prob = 0.65
    energies = []
    days = []
    rise_times = []

    npts = 0
    while npts < maxpts:
        
        e = (2.5*np.random.random() + 0.5) # This is the energy
        t = (max_days)*np.random.random()
        rt = (6.0)*np.random.random()

        prob = 0.0806*e*e + -0.5125*e + 1
    
        rt_slow = rise_time_prob_exp_progression(rt,e,mu,sigma,0,7)
        prob *= rt_slow

        time_prob = np.exp(-0.0002*t)
        
        prob *= time_prob
        probtest = max_prob*np.random.random() # This is to see whether or not we keep x!

        if probtest<prob:
            energies.append(e)
            days.append(t)
            rise_times.append(rt)
            npts += 1
    
    a=energies
    b=days
    c=rise_times
    zip(a,b,c)
    with open(name_of_output_file,'w') as f:
        writer = csv.writer(f, delimiter='\t')
        writer.writerows(zip(a,b,c))
        
    return a,b,c
    
energies,days,rise_times = gen_surface_events(600,1000,'test_data_surface.dat')
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,3,1)
h = plt.hist(energies,bins=50)
plt.subplot(1,3,2)
h = plt.hist(days,bins=50)
plt.subplot(1,3,3)
h = plt.hist(rise_times,bins=50)
```

```python
def gen_flat_events(maxpts,max_days,name_of_output_file):
    
    # Using Nicole's simulated stuff
    mu = [0.269108,0.747275,0.068146]
    sigma = [0.531530,-0.020523]

    # Using Nicole's simulated stuff
    fast_mean_rel_k = [0.431998,-1.525604,-0.024960]
    fast_sigma_rel_k = [-0.014644,5.745791,-6.168695]
    fast_num_rel_k = [-0.261322,5.553102,-5.9144]

    mu0 = [0.374145,0.628990,-1.369876]
    sigma0 = [1.383249,0.495044,0.263360]


    max_prob = 88.2
    energies = []
    days = []
    rise_times = []

    npts = 0
    while npts < maxpts:
        
        e = (2.5*np.random.random() + 0.5) # This is the energy
        t = (max_days)*np.random.random()
        rt = (5.0)*np.random.random()

        ye0 = pdfs.exp(e,-0.0000001,0.5,2.5)
        yrt0 = rise_time_prob_fast_exp_dist(rt,e,mu0,sigma0,fast_mean_rel_k,fast_sigma_rel_k,fast_num_rel_k,0,7)
        yt0 = np.exp(-0.0002*t)

        ye1 = pdfs.exp_plus_flat(e,0.921,17.4,2.38,0.5,2.5)
        yrt1 = rise_time_prob_fast_exp_dist(rt,e,mu0,sigma0,fast_mean_rel_k,fast_sigma_rel_k,fast_num_rel_k,0,7)
        yt1 = np.exp(-0.00000*t)

        nflat = 2287/1000.
        nneutrons = 862/1000.

        prob = nflat*(ye0*yrt0*yt0) + nneutrons*(ye0*yrt0*yt0)
        
        probtest = max_prob*np.random.random() # This is to see whether or not we keep x!

        if probtest<prob:
            energies.append(e)
            days.append(t)
            rise_times.append(rt)
            npts += 1
    
    a=energies
    b=days
    c=rise_times
    zip(a,b,c)
    with open(name_of_output_file,'w') as f:
        writer = csv.writer(f, delimiter='\t')
        writer.writerows(zip(a,b,c))
        
    return a,b,c
    
energies,days,rise_times = gen_flat_events(200,1000,'test_data_flat.dat')
```

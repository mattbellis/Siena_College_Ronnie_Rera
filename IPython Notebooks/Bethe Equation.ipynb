```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# This function plots the stopping power of particles through material modeled by the Bethe formula
# The function takes in Z-atomic number of material, A-atomic mass of material, M-mass of particle passing through, q-charge of particle

def Bethe(Z,A,M,q): 
    Me = 0.510998
    K = 0.307075
    x=[]
    y=[]
    for beta in np.arange(0.01,1,0.0001):
        gamma = 1/(1-(beta**2))
        wmax = 2*Me*beta**2*gamma**2/(1+(2*gamma*Me/(M*9e16))+(Me/(M*9e16))**2)
        dedx = K*q**2*Z/(A*beta**2)*(0.5*np.log((2*Me*beta**2*gamma**2*wmax/((1e-5*Z)**2)))-beta**2)
        momentum = beta*gamma*M
        x.append(momentum)
        y.append(dedx)
    label = r'%5.1f MeV/c$^{2}$' % (M)
    plt.plot(x,y,label=label)
    plt.xscale('log')
    plt.yscale('log')
    plt.ylabel(r'<-dE/dx>(MeV cm$^2$ g$^{-1}$)')
    plt.xlabel('Particle Momentum (MeV/c)')
    plt.title('Passage of Particles Through Matter')
        
Bethe(14,28,105.6,-1)
Bethe(14,28,139,-1)
Bethe(14,28,493,-1)
Bethe(14,28,938.2,+1)

plt.legend()
```

```python
# This function calculates the particles beta after each layer and the time it takes to travel inbetween layers of the CMS tracker

def layer_timing(Z,A,M,q,Momentum): 
    Me = 0.510998
    K = 0.307075
    c = 3e10
    depEnergy = []
    B = []
    layer_sep = [2.9,3.8,14,10,7,8,10,10,8,9,9,13]    #Seperation distance between layers in cm
    energy0 = np.sqrt(Momentum**2+(M)**2)
    beta = Momentum/energy0
    for i in range(0,12):
        gamma = 1/(1-(beta**2))
        wmax = 2*Me*beta**2*gamma**2/(1+(2*gamma*Me/(M*9e16))+(Me/(M*9e16))**2)
        de_pixel_layer = K*q**2*Z/(A*beta**2)*(0.5*np.log((2*Me*beta**2*gamma**2*wmax/((1e-5*Z)**2)))-beta**2)*0.0285*2.33
        de_inner_barrel = K*q**2*Z/(A*beta**2)*(0.5*np.log((2*Me*beta**2*gamma**2*wmax/((1e-5*Z)**2)))-beta**2)*0.032*2.33
        de_outer_barrel = K*q**2*Z/(A*beta**2)*(0.5*np.log((2*Me*beta**2*gamma**2*wmax/((1e-5*Z)**2)))-beta**2)*0.05*2.33
        if i < 3:
            energy_new = energy0 - de_pixel_layer
            depEnergy.append(de_pixel_layer)
        if i >=3 and i < 7:
            energy_new = energy0 - de_inner_barrel
            depEnergy.append(de_inner_barrel)
        if i >= 7 :
            energy_new = energy0 - de_outer_barrel
            depEnergy.append(de_outer_barrel)
            
        momentum_new = np.sqrt(energy_new**2 - M**2)
        beta_new = momentum_new/energy_new
        B.append(beta_new)
        energy0 = energy_new
        Momentum = momentum_new
        beta = beta_new
    #print B
    ttot=0
    for i in range(0,12):
        t = layer_sep[i]/(B[i]*c)
        ttot+=t
    print 'total time(sec):',ttot
    print 'total enery deposited(MeV):',sum(depEnergy)
        
layer_timing(14,28,139,-1,50e3)
layer_timing(14,28,938,-1,50e3)
```

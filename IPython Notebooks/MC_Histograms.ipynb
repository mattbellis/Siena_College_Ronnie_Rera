```python
# This function takes an MC file and creates a histogram of the days, energies and rise times

def MC_hist(FILE1,FILE2,bin_num):
    days1 = FILE1[:,0]
    energies1 = FILE1[:,1]
    RiseTimes1 = FILE1[:,2]
    
    days2 = FILE2[:,0]
    energies2 = FILE2[:,1]
    RiseTimes2 = FILE2[:,2]
    
    index0 = energies2<=3.2
    index1 = energies2>=0.6
    
    index = index0*index1
    index = energies2>-1
    
    plt.figure(figsize=(6,4),dpi=100)
    days_hist = lch.hist_err(days1,bins=bin_num,color='black',ecolor='black',label='CoGeNT data',linewidth=2)
    days_hist = lch.hist_err(days2[index],bins=bin_num,color='red',ecolor='red',label='Monte Carlo',linewidth=2,alpha=0.5)
    plt.xlabel('days since Dec 3 2009',size=18)
    plt.ylabel('Events',size=18)
    plt.legend(fontsize=18)
    plt.xlim(0,1238)
    plt.tight_layout()
    plt.savefig('cogent_MC_comparison_days.png')
    #plt.savefig('wimp_Days.png',bbox_inches='tight')
    
    plt.figure(figsize=(6,4),dpi=100)
    energies_hist = lch.hist_err(energies1,bins=bin_num,color='black',ecolor='black',label='CoGeNT data',range=(0.6,3.2),linewidth=2)
    energies_hist = lch.hist_err(energies2[index],bins=bin_num,color='red',ecolor='red',label='Monte Carlo',range=(0.6,3.2),linewidth=2,alpha=0.5)
    plt.xlabel('energy (keVee)',size=18)
    plt.ylabel('Events',size=18)
    plt.xlim(0.6,3.2)
    plt.legend(fontsize=18)
    plt.tight_layout()
    plt.savefig('cogent_MC_comparison_energy.png')
    #plt.savefig('wimp_Energies.png',bbox_inches='tight')
    
    plt.figure(figsize=(6,4),dpi=100)
    RiseTime_hist = lch.hist_err(RiseTimes1,bins=bin_num,color='black',ecolor='black',label='CoGeNT data',range=(0,6),linewidth=2)
    RiseTime_hist = lch.hist_err(RiseTimes2[index],bins=bin_num,color='red',ecolor='red',label='Monte Carlo',range=(0,6),linewidth=2,alpha=0.5)
    plt.xlabel('Rise time ($\mu$sec)',size=18)
    plt.ylabel('Events',size=18)
    plt.legend(fontsize=18)
    plt.tight_layout()
    plt.savefig('cogent_MC_comparison_rise_time.png')
    #plt.savefig('wimp_Rise_Times.png',bbox_inches='tight')
   

MC_hist(file5,file6,100)
```

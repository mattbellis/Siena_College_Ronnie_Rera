```python
# This function takes an MC file and creates a histogram of the days, energies and rise times

def MC_hist(FILE,bin_num):
    days = FILE[:,0]
    energies = FILE[:,1]
    RiseTimes = FILE[:,2]
    
    plt.figure()
    days_hist = plt.hist(days,bins=bin_num)
    plt.title("Days")
    
    plt.figure()
    energies_hist = plt.hist(energies,bins=bin_num)
    plt.title("Energies")
    
    plt.figure()
    RiseTime_hist = plt.hist(RiseTimes,bins=bin_num)
    plt.title("Rise Times")
    
MC_hist(file0,100)
```

```python
# This function takes an MC file and creates a histogram of the days, energies and rise times

def MC_hist(FILE1,FILE2,bin_num):
    days1 = FILE1[:,0]
    energies1 = FILE1[:,1]
    RiseTimes1 = FILE1[:,2]
    
    days2 = FILE2[:,0]
    energies2 = FILE2[:,1]
    RiseTimes2 = FILE2[:,2]
    
    plt.figure()
    days_hist = lch.hist_err(days1,bins=bin_num,color='black',ecolor='black',label='CoGeNT data')
    days_hist = lch.hist_err(days2,bins=bin_num,color='red',ecolor='red',label='Monte Carlo')
    plt.xlabel('days since Dec 3 2009',size=18)
    plt.ylabel('Events',size=18)
    plt.legend()
    #plt.savefig('wimp_Days.png',bbox_inches='tight')
    
    plt.figure()
    energies_hist = lch.hist_err(energies1,bins=bin_num,color='black',ecolor='black',label='CoGeNT data',range=(0.5,3.2))
    energies_hist = lch.hist_err(energies2,bins=bin_num,color='red',ecolor='red',label='Monte Carlo')
    plt.xlabel('energy (keVee)',size=18)
    plt.ylabel('Events',size=18)
    plt.legend()
    #plt.savefig('wimp_Energies.png',bbox_inches='tight')
    
    plt.figure()
    RiseTime_hist = lch.hist_err(RiseTimes1,bins=bin_num,color='black',ecolor='black',label='CoGeNT data',range=(0,6))
    RiseTime_hist = lch.hist_err(RiseTimes2,bins=bin_num,color='red',ecolor='red',label='Monte Carlo',range=(0,6))
    plt.xlabel('Rise time ($\mu$sec)',size=18)
    plt.ylabel('Events',size=18)
    plt.legend()
    #plt.savefig('wimp_Rise_Times.png',bbox_inches='tight')
   

MC_hist(file5,file6,100)
```

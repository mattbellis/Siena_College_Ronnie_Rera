```python
import numpy as np
import lichen.lichen as lch
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
```

```python
file0 = np.loadtxt('cogent_sample.dat')
#file0 = np.loadtxt('cogent_for_ron.dat')
file1 = np.loadtxt('c4_sample.dat')
```

```python
# This function takes an MC file and creates a histogram of the days, energies and rise times

def MC_hist(FILE1,FILE2,bin_num):
    
    label1 = "1-year CoGeNT MC"
    label2 = "1-year C4 MC"
    
    days1 = FILE1[:,0]
    energies1 = FILE1[:,1]
    RiseTimes1 = FILE1[:,2]
    
    days2 = FILE2[:,0]
    energies2 = FILE2[:,1]
    RiseTimes2 = FILE2[:,2]
    
    index1 =  energies1<=3.2
    index1 *= energies1>=0.5
    
    index1 *= days1<=366

    index2 =  energies2<=3.2
    index2 *= energies2>=0.5
    
    print len(days1)
    print len(days1[index1])

    print len(days2)
    print len(days2[index2])
    
    plt.figure(figsize=(16,4),dpi=100)
    plt.subplot(1,3,1)
    days_hist = lch.hist_err(days1[index1],bins=bin_num,color='black',ecolor='black',label=label1,linewidth=2)
    days_hist = lch.hist_err(days2[index2],bins=bin_num,color='red',ecolor='red',label=label2,linewidth=2,alpha=0.5)
    plt.xlabel('days since Dec 3 2009',size=18)
    plt.ylabel('Events',size=18)
    plt.legend(fontsize=18)
    plt.xlim(0,365)
    plt.tight_layout()
    plt.savefig('cogent_MC_comparison_days.png')
    #plt.savefig('wimp_Days.png',bbox_inches='tight')
    
    plt.subplot(1,3,2)
    energies_hist = lch.hist_err(energies1[index1],bins=bin_num,color='black',ecolor='black',label=label1,range=(0.6,3.2),linewidth=2)
    energies_hist = lch.hist_err(energies2[index2],bins=bin_num,color='red',ecolor='red',label=label2,range=(0.6,3.2),linewidth=2,alpha=0.5)
    plt.xlabel('energy (keVee)',size=18)
    plt.ylabel('Events',size=18)
    plt.xlim(0.5,3.2)
    plt.legend(fontsize=18)
    plt.tight_layout()
    plt.savefig('cogent_MC_comparison_energy.png')
    #plt.savefig('wimp_Energies.png',bbox_inches='tight')
    
    plt.subplot(1,3,3)
    RiseTime_hist = lch.hist_err(RiseTimes1[index1],bins=bin_num,color='black',ecolor='black',label=label1,range=(0,6),linewidth=2)
    RiseTime_hist = lch.hist_err(RiseTimes2[index2],bins=bin_num,color='red',ecolor='red',label=label2,range=(0,6),linewidth=2,alpha=0.5)
    plt.xlabel('Rise time ($\mu$sec)',size=18)
    plt.ylabel('Events',size=18)
    plt.legend(fontsize=18)
    plt.tight_layout()
    plt.savefig('cogent_MC_comparison_rise_time.png')
    #plt.savefig('wimp_Rise_Times.png',bbox_inches='tight')
   

#MC_hist(file5,file6,100)
MC_hist(file0,file1,100)
```

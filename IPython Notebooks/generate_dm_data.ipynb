```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline

import dm_models as dmm
from dm_models import dRdErSHM


#%matplotlib inline
```

```python
target_atom = dmm.AGe # Germanium
mDM = 10 # GeV/c^2
sigma_n = 1e-42 # cross section for DM/nucleon interaction

for day in [0,60,120,180,240,300]:
    # Recoil energy range.
    Er = np.linspace(0.1,10.0,100)

    dRdEr = dRdErSHM(Er,day,target_atom,mDM,sigma_n) # SHM = Simple Halo Model (Maxwellian velocity distribution)
    
    xkeVee = dmm.quench_keVr_to_keVee(Er)
    
    plt.plot(xkeVee,dRdEr)
    
#plt.xlim(0.9,1.1)
#plt.ylim(0.02,0.04)
```

# Your goal, using another set of code, is to generate 10000 events, assuming they were in the CoGeNT detector. So, keVee (energy range from 0.5 - 3 keVee and over the course of 4 years.

```python
maxpts = 10000
xpts = [] # These are the points I ``accept"
max_prob = 0.07
energies = []
days = []

npts = 0
while npts < maxpts:
    
    e = (2.5*np.random.random() + 0.5) # This is the energy
    t = (4*365.0)*np.random.random()
    
    e = dmm.quench_keVee_to_keVr(e)
    
    prob = dRdErSHM(e,t,target_atom,mDM,sigma_n)
    #prob = dmm.quench_keVr_to_keVee(Er)
    
    #print prob
    
    probtest = max_prob*np.random.random() # This is to see whether or not we keep x!
    
    if probtest<prob:
        energies.append(e)
        days.append(t)
        npts += 1
```
